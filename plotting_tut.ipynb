```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(31)

mu = 3
n = 50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
```

```python
plt.rcParams["figure.figsize"] = (10,8)
```

## 3d Scatter Plot

```python
ax = plt.axes(projection="3d")
ax.scatter3D(x, y, z, s=100)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.view_init(10,0)
```

## 3D surface plots


```python
def function_z(x, y):
    return 50 - (x**2 + y**2)
```

```python
N = 50

x_values = np.linspace(-5 , 5 , N)
y_values = np.linspace(-5 , 5 , N)
```

```python
X, Y = np.meshgrid(x_values, y_values)
```

```python
X
```

```python
Y
```

```python
plt.scatter(X, Y)
```

```python
Z = function_z(X, Y)
# Z.shape
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z)
```

## 3D Wireframe

```python
N = 10

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)
```

```python
X, Y = np.meshgrid(x_values, y_values)
Z = function_z(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color="black", lw=3)
ax.plot_wireframe(X, Y, Z, color="green", lw=1)
```
